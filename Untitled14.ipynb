```python
# Cell 2 — Imports
import os
import numpy as np
from rank_bm25 import BM25Okapi
from sentence_transformers import SentenceTransformer, CrossEncoder
import google.generativeai as genai
from groq import Groq

# Set your API keys
GEMINI_API_KEY = "your_key"
GROQ_API_KEY   = "your_key"

genai.configure(api_key=GEMINI_API_KEY)
groq_client = Groq(api_key=GROQ_API_KEY)

# Change this line in Cell 2:
gemini_model = genai.GenerativeModel("gemini-2.0-flash")
```

```python
# Cell 3 — Part 1: Document Corpus (10+ docs)
corpus = [
    # Attention / Transformers
    "The attention mechanism allows transformers to weigh the importance of different words when encoding a sequence.",
    "Self-attention computes a weighted sum of all tokens in a sequence, capturing long-range dependencies.",
    "Multi-head attention runs several attention operations in parallel, each learning different relationships.",

    # Neural Network Training
    "Backpropagation computes gradients by applying the chain rule from the output layer back to the input.",
    "Gradient descent updates model weights by moving in the direction that minimizes the loss function.",
    "Batch normalization stabilizes training by normalizing layer inputs, allowing higher learning rates.",

    # Optimization
    "Adam optimizer combines momentum and adaptive learning rates, making it popular for deep learning.",
    "Learning rate scheduling reduces the learning rate over time to help models converge to better minima.",
    "Dropout randomly deactivates neurons during training, acting as a regularization technique.",

    # Technical jargon doc (good for BM25)
    "The BLEU score (Bilingual Evaluation Understudy) is a precision-based metric for evaluating NLP output quality.",

    # Extra semantic docs
    "Convolutional neural networks (CNNs) use filters to detect spatial features like edges and textures in images.",
    "Recurrent neural networks (RNNs) process sequential data by maintaining a hidden state across time steps.",
    "Transfer learning reuses a pretrained model's weights as a starting point for a new but related task.",
]

print(f"Corpus size: {len(corpus)} documents")
```

```python
# Cell 4 — Part 2: HybridRetriever

class HybridRetriever:
    def __init__(self, corpus: list[str], k: int = 60):
        self.corpus = corpus
        self.k = k  # RRF constant

        # BM25 index (tokenize on whitespace, lowercase)
        tokenized = [doc.lower().split() for doc in corpus]
        self.bm25 = BM25Okapi(tokenized)

        # SBERT index
        self.sbert = SentenceTransformer("all-MiniLM-L6-v2")
        self.doc_embeddings = self.sbert.encode(corpus, convert_to_numpy=True)

    def retrieve(self, query: str, top_k: int = 5) -> list[dict]:
        # --- BM25 ---
        bm25_scores = self.bm25.get_scores(query.lower().split())
        bm25_ranked = np.argsort(bm25_scores)[::-1]  # descending

        # --- SBERT cosine similarity ---
        q_emb = self.sbert.encode([query], convert_to_numpy=True)
        sbert_scores = (self.doc_embeddings @ q_emb.T).squeeze()
        sbert_ranked = np.argsort(sbert_scores)[::-1]  # descending

        # --- RRF fusion ---
        # Build rank lookup dicts {doc_id: rank (1-indexed)}
        bm25_rank_map  = {doc_id: rank+1 for rank, doc_id in enumerate(bm25_ranked)}
        sbert_rank_map = {doc_id: rank+1 for rank, doc_id in enumerate(sbert_ranked)}

        rrf_scores = {}
        for doc_id in range(len(self.corpus)):
            r_bm25  = bm25_rank_map.get(doc_id, len(self.corpus))
            r_sbert = sbert_rank_map.get(doc_id, len(self.corpus))
            rrf_scores[doc_id] = (1 / (self.k + r_bm25)) + (1 / (self.k + r_sbert))

        # Sort by RRF score descending
        ranked_ids = sorted(rrf_scores, key=rrf_scores.get, reverse=True)[:top_k]

        results = []
        for doc_id in ranked_ids:
            results.append({
                "doc_id":     doc_id,
                "rrf_score":  rrf_scores[doc_id],
                "bm25_rank":  bm25_rank_map[doc_id],
                "sbert_rank": sbert_rank_map[doc_id],
                "text":       self.corpus[doc_id],
            })
        return results

# Test it
retriever = HybridRetriever(corpus)
results = retriever.retrieve("how does attention work?", top_k=5)
for r in results:
    print(f"[RRF={r['rrf_score']:.4f} | BM25 rank={r['bm25_rank']} | SBERT rank={r['sbert_rank']}]")
    print(f"  {r['text']}\n")
```

```python
# Cell 5 — Part 3: Cross-Encoder Re-Ranker

cross_encoder = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

def rerank(query: str, candidates: list[dict], top_k: int = 3) -> list[dict]:
    """Re-rank candidates using cross-encoder. Uses original user query."""
    pairs = [(query, c["text"]) for c in candidates]
    scores = cross_encoder.predict(pairs)

    for i, score in enumerate(scores):
        candidates[i]["cross_encoder_score"] = float(score)

    reranked = sorted(candidates, key=lambda x: x["cross_encoder_score"], reverse=True)
    return reranked[:top_k]

# Test it
query = "how do transformers encode meaning?"
candidates = retriever.retrieve(query, top_k=5)
top_docs = rerank(query, candidates, top_k=3)

print("After re-ranking:")
for d in top_docs:
    print(f"  [CE score={d['cross_encoder_score']:.3f}] {d['text']}")
```

```python
def hyde_expand(user_query: str) -> str:
    """Generate a hypothetical answer using Groq."""
    prompt = f"""Write a short, factual 2-sentence answer to this question as if you were an AI textbook.
Question: {user_query}
Answer:"""

    response = groq_client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.0,
    )
    return response.choices[0].message.content.strip()

# Test HyDE
expanded = hyde_expand("what is attention?")
print("Original query: what is attention?")
print(f"HyDE expansion: {expanded}")
```

```python
def advanced_rag(user_query: str) -> str:
    """
    Full pipeline:
    1. Query Expansion (HyDE)
    2. Hybrid Retrieval (BM25 + SBERT + RRF)
    3. Re-Ranking (Cross-Encoder) — uses ORIGINAL query
    4. LLM Generation (Groq)
    """
    print(f"\n=== Query: {user_query} ===")

    # Step 1 — HyDE expansion
    expanded_query = hyde_expand(user_query)
    print(f"[HyDE] Expanded: {expanded_query[:100]}...")

    # Step 2 — Hybrid retrieval using expanded query
    candidates = retriever.retrieve(expanded_query, top_k=5)
    print(f"[Retrieval] Retrieved {len(candidates)} candidates")

    # Step 3 — Re-rank using ORIGINAL user query
    top_docs = rerank(user_query, candidates, top_k=3)
    context = "\n".join([d["text"] for d in top_docs])
    print(f"[Re-rank] Top doc: {top_docs[0]['text'][:80]}...")

    # Step 4 — Generate answer with Groq
    prompt = f"""You are a helpful university AI assistant. Answer the student's question using ONLY the context below.

Context:
{context}

Question: {user_query}
Answer:"""

    response = groq_client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2,
    )
    return response.choices[0].message.content.strip()

# Test full pipeline
answer = advanced_rag("how do transformers encode meaning?")
print(f"\nAnswer: {answer}")
```

```python
def naive_rag(user_query: str) -> str:
    """Dense-only retrieval (SBERT cosine), no expansion, no re-ranking."""
    q_emb = retriever.sbert.encode([user_query], convert_to_numpy=True)
    scores = (retriever.doc_embeddings @ q_emb.T).squeeze()
    top_idx = int(np.argmax(scores))
    top_doc = corpus[top_idx]

    prompt = f"""Answer this question using only the context below.
Context: {top_doc}
Question: {user_query}
Answer:"""

    response = groq_client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2,
    )
    return response.choices[0].message.content.strip()
```

```python
test_queries = [
    "how do transformers encode meaning?",
    "optimization techniques for training",
    "what is the BLEU score used for?",
]

for q in test_queries:
    print(f"\n{'='*60}")
    print(f"Query: {q}")

    # Naïve RAG top doc
    q_emb = retriever.sbert.encode([q], convert_to_numpy=True)
    scores = (retriever.doc_embeddings @ q_emb.T).squeeze()
    naive_top = corpus[int(np.argmax(scores))]

    # Advanced RAG top doc
    expanded = hyde_expand(q)
    candidates = retriever.retrieve(expanded, top_k=5)
    adv_top = rerank(q, candidates, top_k=1)[0]["text"]

    different = "Yes" if naive_top != adv_top else "No"
    print(f"Naive top:    {naive_top[:80]}")
    print(f"Advanced top: {adv_top[:80]}")
    print(f"Different?    {different}")
```

## What I noticed

Honestly the results were a mix. For the transformers question, both pipelines
picked the exact same document, which makes sense — the query was already
specific enough that even basic semantic search knew what to look for.
HyDE didn't really add anything extra here.

The optimization query was where things got interesting. Naïve RAG went with
learning rate scheduling, which is a reasonable answer but pretty surface-level.
The advanced pipeline came back with batch normalization instead — which is
arguably a more insightful answer to the question. I think the HyDE expansion
helped here because it turned a vague 4-word query into something the retriever
could actually work with.

The BLEU score question was an easy win for both. The word "BLEU" is so specific
that even keyword matching nails it immediately — no fancy expansion needed.

So overall, advanced RAG didn't always win, but when the query was vague it
clearly did better. That's kind of the whole point — it's built for the hard cases.
